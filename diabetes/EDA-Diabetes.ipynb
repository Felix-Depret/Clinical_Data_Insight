### <a id='2.3'>2.3 Description des variables

Pregnancies
Nombre de fois enceinte
Nous avons constaté que les femmes ayant eu ≥ 4 grossesses avaient une glycémie plasmatique à jeun (FPG), une hémoglobine glyquée (HbA1c), une glycémie sur 2 heures et un taux de glycémie à jeun (FPG) significativement plus élevées que celles ayant eu deux à trois grossesses (toutes P < 0,01). . Ces femmes avaient également une prévalence de diabète significativement plus élevée (28,4 % contre 20,7 % ; P < 0,001). En utilisant le groupe de deux à trois grossesses comme référence, nous avons observé une association positive du log-FPG et du log-HbA1c avec 4 grossesses ou plus après ajustement pour tenir compte des facteurs sociodémographiques, du mode de vie et de la reproduction, ainsi que de l'indice de masse corporelle (tous deux P < 0,05). . Par rapport aux femmes ayant eu deux ou trois grossesses, les rapports de cotes pour le diabète étaient de 1,31 (intervalle de confiance [IC] à 95 % 1,01-1,71) pour les femmes qui n'étaient jamais tombées enceintes et de 1,28 (IC à 95 % 1,10-1,48) pour celles qui avaient eu au moins 4 grossesses. grossesses après ajustement multivarié. 
Une autre étude de l'étude Healthy Baby Cohort a révélé que les femmes ayant eu ≥3 grossesses avaient un risque 1,27 fois plus élevé de développer un diabète sucré gestationnel (DG) que celles ayant eu moins de grossesses. L'étude suggère qu'un nombre plus élevé de grossesses est associé à une prévalence accrue du diabète gestationnel.

Glucose
Concentration de glucose plasmatique à 2 heures lors d'un test de tolérance au glucose oral
Un test de glycémie postprandial de 2 heures, également connu sous le nom de test de glycémie, postprandial ou PPG de 2 heures, est un test sanguin utilisé pour vérifier le diabète. Il mesure la façon dont votre corps réagit au sucre et à l'amidon après un repas. Le test est effectué en prélevant un échantillon de sang 2 heures après avoir mangé, pour voir si votre glycémie est revenue à la normale. Les résultats normaux pour les personnes non diabétiques sont généralement inférieurs à 140 mg/dL et pour les personnes diabétiques, inférieurs à 180 mg/dL.
Le test est souvent utilisé pour diagnostiquer le diabète et surveiller le contrôle de la glycémie chez les diabétiques connus.

Blood Pressure
Pression artérielle diastolique (mm Hg)	
Les personnes atteintes de diabète courent un risque plus élevé de développer une hypertension artérielle, également appelée hypertension. L'hypertension artérielle peut entraîner diverses complications chez les personnes atteintes de diabète, telles que des maladies cardiaques, des accidents vasculaires cérébraux, des maladies oculaires diabétiques et des maladies rénales. L'objectif de tension artérielle recommandé pour les personnes diabétiques est généralement inférieur à 140/90 mmHg, ou inférieur à 150/90 mmHg pour les personnes âgées de 80 ans ou plus.

Skin Thickness
Épaisseur du pli cutané du triceps (mm)
L'épaisseur de la peau est principalement déterminée par la teneur en collagène et augmente dans le diabète sucré insulino-dépendant (IDDM). L'épaisseur de la peau a été examinée chez 66 patients atteints de DSID âgés de 24 à 38 ans. L'épaisseur de la peau était augmentée et significativement liée à la durée du diabète

Insulin
Insuline sérique à 2 heures (mu U/ml)
Le taux d'insuline sérique 2 heures après une charge de glucose est un indicateur important de la résistance à l'insuline. 
La plage de référence pour les taux d'insuline 2 heures après l'administration du glucose est de 16 à 166 mUI/L ou de 111 à 1 153 pmol/L.

BMI
Body Mass Index
Indice de masse corporelle (poids en kg/(taille en m)^2)	L'indice de masse corporelle (IMC) est une mesure de la graisse corporelle basée sur le poids et la taille d'une personne. Il est couramment utilisé pour évaluer si un individu présente une insuffisance pondérale, un poids normal, un surpoids ou une obésité. La recherche a montré qu'il existe une relation étroite entre l'IMC et le risque de développer un diabète. Par exemple, les femmes ayant un IMC de 30 kg/m2 ont un risque 28 fois plus élevé de développer un diabète que les femmes de poids normal, et le risque augmente jusqu'à 93 fois plus avec un IMC de 35 kg/m2.

Diabetes Pedigree Function
Une fonction qui évalue la probabilité de diabète en fonction des antécédents familiaux
La fonction généalogique du diabète est une fonction mathématique qui évalue la probabilité de diabète en fonction des antécédents familiaux et de la relation génétique entre les proches et le patient. Il fournit des données sur les antécédents de diabète sucré chez les proches et leur relation génétique avec le patient, donnant une idée du risque héréditaire que l'on peut avoir avec l'apparition du diabète sucré.
Cette fonction est utilisée comme prédicteur dans le contexte de l'apprentissage automatique et de la modélisation statistique pour estimer la présence du diabète, et elle s'est avérée positivement associée au diabète.
C'est l'un des prédicteurs importants dans la prédiction du diabète de type 2, avec d'autres facteurs tels que le taux de glucose, l'IMC et l'âge.

Âge
Âge en années
Le lien entre l’âge et le diabète est bien établi, l’âge étant un facteur de risque important pour le développement du diabète de type 2. Des recherches ont montré que la prévalence du diabète et du prédiabète augmente avec l'âge, avec un risque plus élevé chez les personnes âgées que chez la population d'âge moyen.

```python
def count_unique_values(dataframe):
    unique_counts_df = dataframe.nunique().to_frame()
    unique_counts_df.reset_index(inplace=True)
    unique_counts_df.rename(columns={'index': 'column', 0: 'nb_unique_values'}, inplace=True)
    return unique_counts_df
```

### <a id='3.6'>3.6 .isna().sum()

```python
def count_isna(df):
    isna_df= df.isna().sum().to_frame()
    isna_df.reset_index(inplace = True)
    isna_df.rename(columns={'index':'column', 0: 'nb_isna'}, inplace = True)
    isna_df= isna_df.loc[(isna_df != 0).all(axis=1)]
    return isna_df
```

```python
count_isna(df_diabetes)
```

```python
def count_iszero(df):
    df_iszero = pd.DataFrame(df[df == 0].count())
    df_iszero.rename(columns={0: 'nb_iszero'}, inplace = True)
    df_iszero = df_iszero.loc[~(df_iszero == 0).all(axis=1)]
    return df_iszero
```

On va observer la distribution après remplacement des 0 par plusieurs méthodes pour définir la méthode de remplacement.

```python
def impute_and_plot_columns_with_0(data):
    data_imputed = data.copy()
    for col in data.columns:
         # Calculer la moyenne et la médiane de la colonne
        col_mean = data_imputed[col].mean()
        col_median = data_imputed[col].median()
        
        data_imputed[col].replace(0,  np.nan, inplace = True)
        
        
                # Imputer les valeurs manquantes avec la moyenne, la médiane et la méthode de remplissage avant
        data_imputed[f'{col}_mean'] = data_imputed[col].fillna(col_mean)
        data_imputed[f'{col}_median'] = data_imputed[col].fillna(col_median)
        data_imputed[f'{col}_ffill'] = data_imputed[col].fillna(method='ffill')
        data_imputed[f'{col}_bfill'] = data_imputed[col].fillna(method='bfill')
        
                # Tracer les distributions résultantes
        plt.figure(figsize=(12, 6))
        sns.kdeplot(data_imputed[f'{col}_mean'], color='red', label='Mean')
        sns.kdeplot(data_imputed[f'{col}_median'], color='blue', label='Median')
        sns.kdeplot(data_imputed[f'{col}_ffill'], color='yellow', label='ffill')
        sns.kdeplot(data_imputed[f'{col}_bfill'], color='green', label='bfill')
        sns.kdeplot(data_imputed[col], color='black', label='Original')
        plt.title(f'Distribution de la colonne {col}')
        plt.legend()
        plt.show()
    print(count_iszero(data_imputed))
    print(count_iszero(data))
        
```

```python
impute_and_plot_columns_with_0(df_diabetes)
```

```python
def fill_0_mean(dataset, colonne_a_modifier, colonne_categorie):
     # Séparer les lignes malades et saines
    classe0 = dataset[dataset[colonne_categorie] == 0]
    classe1 = dataset[dataset[colonne_categorie] == 1]
    moyenne_classe0 = classe0[colonne_a_modifier].mean()
    moyenne_classe1 = classe1[colonne_a_modifier].mean()
    
    dataset[colonne_a_modifier].replace(0,  np.nan, inplace = True)    
    dataset[colonne_a_modifier].fillna(method='bfill', inplace=True)
    # Remplacer les NaN par la moyenne correspondante en fonction de la catégorie du patient
    dataset.loc[dataset[colonne_categorie] == 'classe0', colonne_a_modifier] = \
                dataset.loc[dataset[colonne_categorie] == 'classe0', colonne_a_modifier].fillna(moyenne_classe0)

    dataset.loc[dataset[colonne_categorie] == 'classe1', colonne_a_modifier] = \
                dataset.loc[dataset[colonne_categorie] == 'classe1', colonne_a_modifier].fillna(moyenne_classe1)
    return dataset
```

```python
def boxplot_for_each_column(dataframe, column_diagnostic):
    dataframe[column_diagnostic] = dataframe[column_diagnostic].astype(str)
    df_y= dataframe.drop([column_diagnostic], axis=1)
    df_x= dataframe[column_diagnostic]
    plt.figure(figsize=(15, 8))
    for column in df_y.columns:
        plt.subplot(1,len(df_y.columns), df_y.columns.get_loc(column) + 1)
        ax = sns.boxplot(y=df_y[column], x= df_x)
        add_stat_annotation(ax, data=dataframe, x= column_diagnostic, y=df_y[column],
                    box_pairs=[("0", "1")],
                    test='Kruskal', text_format='full', loc='inside', verbose=2)
        plt.title(column)
    plt.tight_layout()
    plt.show()
    dataframe[column_diagnostic] = dataframe[column_diagnostic].astype(int)
```

```python
boxplot_for_each_column(df_diabetes, 'Outcome')
```

#### <a id='5.3.2'>5.3.2. Heatmaps

```python
def plot_triangle_corr_heatmap(dataframe):
    # Calculer la matrice de corrélation
    corr_matrix = dataframe.corr()

    # Créer un masque pour le triangle supérieur
    mask = np.triu(np.ones_like(corr_matrix))

    # Tracer la heatmap avec seaborn
    plt.figure(figsize=(20, 8))
    heatmap = sns.heatmap(corr_matrix, mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG', fmt='.2f')
    heatmap.set_title('Triangle Correlation Heatmap', fontdict={'fontsize': 18}, pad=16)

    # Afficher la heatmap
    plt.show()
```

```python
plot_triangle_corr_heatmap(df_diabetes.drop(columns= ['Outcome']))
```
